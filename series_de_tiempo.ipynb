```python
# -*- coding: utf-8 -*-
"""
Análisis de Series de Tiempo para Ventas de Videojuegos
Autores: García Rodríguez Héctor Alenkar
Fecha: Junio 2025
"""

# ======================== CONFIGURACIÓN INICIAL ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.diagnostic import acorr_ljungbox
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error
import os

# Configuración global
plt.style.use('seaborn-v0_8-darkgrid')
pd.set_option('display.float_format', '{:.2f}'.format)
np.random.seed(42)
sns.set_palette("viridis")

# Crear directorio para resultados
os.makedirs('resultados', exist_ok=True)

# ======================== CARGA Y PREPROCESAMIENTO ===========================
def cargar_y_preparar_datos():
    """Carga y prepara el dataset de ventas de videojuegos"""
    # Cargar datos (reemplazar con tu ruta real)
    datos = pd.read_csv('vgsales.csv')

    print("\n[INFO] Exploración inicial:")
    print(f"- Filas: {datos.shape[0]}, Columnas: {datos.shape[1]}")
    print(f"- Valores faltantes: {datos.isnull().sum().sum()}")

    # Limpieza básica
    datos = datos.dropna(subset=['Year', 'Global_Sales'])
    datos['Year'] = datos['Year'].astype(int)

    # Filtrar años
    datos = datos[(datos['Year'] >= 1980) & (datos['Year'] <= 2020)]

    # Agregación anual robusta
    ventas_anuales = datos.groupby('Year').agg({
        'Global_Sales': 'mean',
        'NA_Sales': 'mean',
        'EU_Sales': 'mean',
        'JP_Sales': 'mean',
        'Other_Sales': 'mean',
        'Name': 'count'  # Conteo de juegos por año
    }).rename(columns={'Name': 'Num_Juegos'}).reset_index()

    # Calcular tasa de crecimiento
    ventas_anuales['Crecimiento'] = ventas_anuales['Global_Sales'].pct_change() * 100

    print("\n[INFO] Datos anuales preparados:")
    print(ventas_anuales.head())

    return ventas_anuales

# ======================== ANÁLISIS EXPLORATORIO ==============================
def analisis_exploratorio(df):
    """Realiza análisis exploratorio visual y estadístico"""
    print("\n[FASE] Análisis Exploratorio")

    # 1. Evolución temporal de ventas
    fig, ax = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

    # Ventas globales
    sns.lineplot(x='Year', y='Global_Sales', data=df, marker='o', ax=ax[0])
    ax[0].set_title('Evolución de Ventas Globales Promedio (1980-2020)', fontsize=14)
    ax[0].set_ylabel('Ventas (millones)', fontsize=12)
    ax[0].grid(True, alpha=0.3)

    # Tasa de crecimiento
    sns.barplot(x='Year', y='Crecimiento', data=df, ax=ax[1], color='skyblue')
    ax[1].axhline(y=0, color='red', linestyle='--', alpha=0.7)
    ax[1].set_title('Tasa de Crecimiento Anual (%)', fontsize=14)
    ax[1].set_ylabel('Crecimiento %', fontsize=12)
    ax[1].set_xlabel('Año', fontsize=12)
    ax[1].tick_params(axis='x', rotation=45)
    ax[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('resultados/evolucion_ventas.png', dpi=300)
    plt.show()

    # 2. Descomposición temporal
    df_temp = df.set_index('Year')
    descomposicion = seasonal_decompose(df_temp['Global_Sales'], period=5, model='additive')

    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(14, 10), sharex=True)

    descomposicion.observed.plot(ax=ax1, title='Observado')
    descomposicion.trend.plot(ax=ax2, title='Tendencia')
    descomposicion.seasonal.plot(ax=ax3, title='Estacionalidad')
    descomposicion.resid.plot(ax=ax4, title='Residuales')

    for ax in [ax1, ax2, ax3, ax4]:
        ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('resultados/descomposicion_temporal.png', dpi=300)
    plt.show()

    # 3. Relación entre variables
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", ax=ax)
    ax.set_title('Matriz de Correlación', fontsize=14)
    plt.tight_layout()
    plt.savefig('resultados/matriz_correlacion.png', dpi=300)
    plt.show()

    return df.set_index('Year')

# ======================== ANÁLISIS DE ESTACIONARIEDAD ========================
def analizar_estacionariedad(serie):
    """Realiza pruebas de estacionariedad y diferenciación"""
    print("\n[FASE] Análisis de Estacionariedad")

    # 1. Pruebas estadísticas
    def realizar_pruebas(s, nombre):
        """Realiza pruebas ADF y KPSS"""
        # Prueba ADF
        adf_result = adfuller(s.dropna())
        # Prueba KPSS
        kpss_result = kpss(s.dropna(), regression='c')

        resultados = {
            'Serie': nombre,
            'ADF_Estadistico': adf_result[0],
            'ADF_pvalor': adf_result[1],
            'KPSS_Estadistico': kpss_result[0],
            'KPSS_pvalor': kpss_result[1]
        }

        # Interpretación
        print(f"\nResultados para {nombre}:")
        print(f"- ADF p-valor: {adf_result[1]:.4f} -> {'Estacionaria' if adf_result[1] < 0.05 else 'No Estacionaria'}")
        print(f"- KPSS p-valor: {kpss_result[1]:.4f} -> {'Estacionaria' if kpss_result[1] > 0.05 else 'No Estacionaria'}")

        return resultados

    resultados = []
    resultados.append(realizar_pruebas(serie, 'Serie Original'))

    # 2. Diferenciación si es necesario
    if adfuller(serie.dropna())[1] > 0.05:
        serie_diff = serie.diff().dropna()
        resultados.append(realizar_pruebas(serie_diff, 'Serie Diferenciada (1)'))
        print("\n[CONCLUSIÓN] Serie requiere diferenciación (d=1)")
        return serie_diff, 1
    else:
        print("\n[CONCLUSIÓN] Serie es estacionaria (d=0)")
        return serie, 0

# ======================== IDENTIFICACIÓN DE MODELO ===========================
def identificar_modelo(serie):
    """Identifica los parámetros óptimos para el modelo ARIMA"""
    print("\n[FASE] Identificación de Modelo ARIMA")

    # 1. Gráficos ACF y PACF
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

    plot_acf(serie, lags=15, ax=ax1, color='#1f77b4', alpha=0.7,
             title='Función de Autocorrelación (ACF)')
    plot_pacf(serie, lags=15, ax=ax2, color='#ff7f0e', alpha=0.7,
              title='Función de Autocorrelación Parcial (PACF)')

    for ax in [ax1, ax2]:
        ax.grid(True, alpha=0.3)
        ax.axhline(y=0, color='black', linestyle='-')
        ax.axhline(y=1.96/np.sqrt(len(serie)), color='red', linestyle='--', alpha=0.7)
        ax.axhline(y=-1.96/np.sqrt(len(serie)), color='red', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.savefig('resultados/correlogramas.png', dpi=300)
    plt.show()

    # 2. Búsqueda en grid de parámetros
    print("\n[INFO] Realizando búsqueda de mejores parámetros...")
    best_aic = np.inf
    best_order = (0, 0, 0)
    resultados_busqueda = []

    # Rango de parámetros a evaluar
    for p in range(0, 3):  # AR
        for d in range(0, 2):  # Diferenciación (ya aplicada)
            for q in range(0, 3):  # MA
                try:
                    modelo = ARIMA(serie, order=(p, 0, q))  # d=0 porque ya diferenciamos
                    resultado = modelo.fit()
                    aic = resultado.aic

                    resultados_busqueda.append({
                        'p': p,
                        'd': d,
                        'q': q,
                        'AIC': aic,
                        'BIC': resultado.bic
                    })

                    if aic < best_aic:
                        best_aic = aic
                        best_order = (p, 0, q)  # d=0 porque ya diferenciamos

                    print(f"ARIMA({p},{d},{q}) - AIC: {aic:.2f}")
                except Exception as e:
                    print(f"Error con ARIMA({p},{d},{q}): {str(e)}")
                    continue

    # Resultados de la búsqueda
    resultados_df = pd.DataFrame(resultados_busqueda)
    resultados_df = resultados_df.sort_values('AIC')
    print("\nTop 5 modelos por AIC:")
    print(resultados_df.head(5).to_string(index=False))

    print(f"\n[CONCLUSIÓN] Mejor modelo: ARIMA{best_order} con AIC={best_aic:.2f}")
    return best_order

# ======================== MODELADO Y VALIDACIÓN ==============================
def modelar_y_validar(serie, orden, test_size=0.2):
    """Entrena, valida y evalúa el modelo ARIMA"""
    print("\n[FASE] Modelado y Validación")

    # 1. División de datos
    n = len(serie)
    split_idx = int(n * (1 - test_size))
    train = serie.iloc[:split_idx]
    test = serie.iloc[split_idx:]

    print(f"- Entrenamiento: {len(train)} años ({train.index.min()}-{train.index.max()})")
    print(f"- Validación: {len(test)} años ({test.index.min()}-{test.index.max()})")

    # 2. Entrenamiento del modelo
    modelo = ARIMA(train, order=orden)
    modelo_fit = modelo.fit()
    print("\nResumen del Modelo:")
    print(modelo_fit.summary())

    # 3. Pronóstico en validación
    pronostico = modelo_fit.get_forecast(steps=len(test), conf_int=True)
    pred = pronostico.predicted_mean
    pred_ci = pronostico.conf_int()

    # 4. Métricas de error
    rmse = np.sqrt(mean_squared_error(test, pred))
    mape = mean_absolute_percentage_error(test, pred) * 100

    print("\n[RESULTADOS] Métricas de Error:")
    print(f"- RMSE: {rmse:.2f}")
    print(f"- MAPE: {mape:.2f}%")

    # 5. Gráfico de pronóstico
    plt.figure(figsize=(14, 7))

    # Histórico
    plt.plot(train.index, train, 'b-o', label='Entrenamiento')

    # Validación
    plt.plot(test.index, test, 'g-o', label='Real', linewidth=2)
    plt.plot(test.index, pred, 'r--o', label='Pronóstico')
    plt.fill_between(test.index, pred_ci.iloc[:, 0], pred_ci.iloc[:, 1], color='gray', alpha=0.2)

    plt.title(f'Pronóstico de Ventas Globales (ARIMA{orden})', fontsize=16)
    plt.xlabel('Año', fontsize=14)
    plt.ylabel('Ventas (millones)', fontsize=14)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.savefig('resultados/pronostico_validacion.png', dpi=300)
    plt.show()

    # 6. Análisis de residuales
    print("\n[ANÁLISIS] Validación de Residuales:")
    residuos = modelo_fit.resid

    # Prueba Ljung-Box
    lb_test = acorr_ljungbox(residuos, lags=[10])
    lb_pvalue = lb_test.lb_pvalue.values[0]
    print(f"- Ljung-Box (p-valor): {lb_pvalue:.4f} -> {'Ruido blanco' if lb_pvalue > 0.05 else 'Autocorrelación'}")

    # Gráficos residuales
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

    # Histograma
    sns.histplot(residuos, kde=True, ax=ax1, color='skyblue')
    ax1.set_title('Distribución de Residuales')
    ax1.axvline(x=0, color='r', linestyle='--')

    # Residuales vs Tiempo
    ax2.scatter(residuos.index, residuos, color='purple', alpha=0.7)
    ax2.axhline(y=0, color='r', linestyle='-')
    ax2.set_title('Residuales vs. Tiempo')
    ax2.set_xlabel('Índice Temporal')
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('resultados/analisis_residuales.png', dpi=300)
    plt.show()

    return modelo_fit, rmse, mape

# ======================== PRONÓSTICO A FUTURO ===============================
def pronosticar_futuro(modelo, serie, pasos=5):
    """Genera pronósticos futuros con intervalos de confianza"""
    print("\n[FASE] Pronóstico a Futuro")

    # 1. Generar pronóstico
    pronostico = modelo.get_forecast(steps=pasos, conf_int=True)
    futuro_pred = pronostico.predicted_mean
    futuro_ci = pronostico.conf_int()

    # 2. Crear índice de años futuros
    ultimo_anio = serie.index.max()
    anos_futuro = range(ultimo_anio + 1, ultimo_anio + pasos + 1)

    # 3. Resultados
    print("\nPronóstico 2021-2025:")
    print(f"{'Año':<6} {'Pronóstico':<12} {'L. Inferior':<12} {'L. Superior':<12}")
    for i, anio in enumerate(anos_futuro):
        print(f"{anio:<6} {futuro_pred[i]:<12.2f} {futuro_ci.iloc[i, 0]:<12.2f} {futuro_ci.iloc[i, 1]:<12.2f}")

    # 4. Gráfico final
    plt.figure(figsize=(14, 8))

    # Histórico
    plt.plot(serie.index, serie, 'b-o', label='Histórico')

    # Futuro
    plt.plot(anos_futuro, futuro_pred, 'r--o', label='Pronóstico')
    plt.fill_between(anos_futuro, futuro_ci.iloc[:, 0], futuro_ci.iloc[:, 1], color='gray', alpha=0.2)

    plt.title('Pronóstico de Ventas Globales 2021-2025', fontsize=16)
    plt.xlabel('Año', fontsize=14)
    plt.ylabel('Ventas (millones)', fontsize=14)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.axvline(x=2020, color='g', linestyle='--', alpha=0.7, label='Fin datos históricos')

    # Formatear ejes
    plt.gca().xaxis.set_major_locator(mdates.YearLocator(5))
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.savefig('resultados/pronostico_futuro.png', dpi=300)
    plt.show()

    # 5. Exportar resultados
    resultados_finales = pd.DataFrame({
        'Año': anos_futuro,
        'Pronostico': futuro_pred.values,
        'Limite_Inferior': futuro_ci.iloc[:, 0].values,
        'Limite_Superior': futuro_ci.iloc[:, 1].values
    })
    resultados_finales.to_csv('resultados/pronostico_futuro.csv', index=False)

    return resultados_finales

# ======================== ANÁLISIS DE SENSIBILIDAD ==========================
def analisis_sensibilidad(serie, orden):
    """Evalúa la robustez del modelo con diferentes splits"""
    print("\n[FASE] Análisis de Sensibilidad")

    resultados = []
    splits = [0.65, 0.70, 0.75, 0.80, 0.85]

    for split in splits:
        print(f"\n- Probando split: {split:.0%} entrenamiento")
        try:
            _, rmse, mape = modelar_y_validar(serie, orden, test_size=1-split)
            resultados.append({
                'Train Split': split,
                'RMSE': rmse,
                'MAPE': mape
            })
        except Exception as e:
            print(f"Error con split {split}: {str(e)}")
            continue

    resultados_df = pd.DataFrame(resultados)
    print("\nResultados de Sensibilidad:")
    print(resultados_df.to_string(index=False))

    # Gráfico de resultados
    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    sns.barplot(x='Train Split', y='RMSE', data=resultados_df, palette='viridis')
    plt.title('RMSE por Proporción de Entrenamiento')
    plt.ylabel('RMSE')

    plt.subplot(1, 2, 2)
    sns.barplot(x='Train Split', y='MAPE', data=resultados_df, palette='magma')
    plt.title('MAPE por Proporción de Entrenamiento')
    plt.ylabel('MAPE (%)')

    plt.tight_layout()
    plt.savefig('resultados/sensibilidad_modelo.png', dpi=300)
    plt.show()

    return resultados_df

# ======================== FLUJO PRINCIPAL ===================================
if __name__ == "__main__":
    print("\n" + "="*80)
    print(" ANÁLISIS DE SERIES DE TIEMPO PARA VENTAS DE VIDEOJUEGOS")
    print("="*80)

    try:
        # 1. Carga y preparación de datos
        datos_anuales = cargar_y_preparar_datos()

        # 2. Análisis exploratorio
        df_ventas = analisis_exploratorio(datos_anuales)
        serie_original = df_ventas['Global_Sales']

        # 3. Análisis de estacionariedad
        serie_estacionaria, d_value = analizar_estacionariedad(serie_original)

        # 4. Identificación de modelo
        orden_arima = identificar_modelo(serie_estacionaria)

        # 5. Modelado y validación
        modelo_final, rmse_final, mape_final = modelar_y_validar(serie_estacionaria, orden_arima)

        # 6. Análisis de sensibilidad
        resultados_sensibilidad = analisis_sensibilidad(serie_estacionaria, orden_arima)

        # 7. Pronóstico a futuro
        pronostico_futuro = pronosticar_futuro(modelo_final, serie_original)

        print("\n" + "="*80)
        print(" ¡ANÁLISIS COMPLETADO EXITOSAMENTE!")
        print("="*80)
        print(f"Resultados guardados en carpeta 'resultados/'")
        print(f"Error final del modelo: MAPE = {mape_final:.2f}%")

    except Exception as e:
        print("\n[ERROR] Ocurrió un problema en el análisis:")
        print(str(e))
        print("\n" + "="*80)
        print(" ANÁLISIS INTERRUMPIDO")
        print("="*80)
```
